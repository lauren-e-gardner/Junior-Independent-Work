```python
## for data
import pandas as pd
import numpy as np
import io
from sklearn.model_selection import train_test_split
## for plotting
import matplotlib.pyplot as plt
import seaborn as sns

# Prompts to upload csv file and uploads it to pandas dataframe
from google.colab import files
uploaded = files.upload()

# Read in data set to pandas
dtf = pd.read_csv('modelnet10.csv')

# Remove split row
dtf.drop(columns='split')

classes = ['bathtub', 'bed', 'chair', 'desk', 'dresser', 'monitor', 'night', 'sofa', 'table', 'toilet']

# Split the data set up by item
dtf_bathtub = dtf[dtf['class'] == 'bathtub']
dtf_bed = dtf[dtf['class'] == 'bed']
dtf_chair = dtf[dtf['class'] == 'chair']
dtf_desk = dtf[dtf['class'] == 'desk']
dtf_dresser = dtf[dtf['class'] == 'dresser']
dtf_monitor = dtf[dtf['class'] == 'monitor']
dtf_night = dtf[dtf['class'] == 'night']
dtf_sofa = dtf[dtf['class'] == 'sofa']
dtf_table = dtf[dtf['class'] == 'table']
dtf_toilet = dtf[dtf['class'] == 'toilet']


frames = [dtf_bathtub, dtf_bed, dtf_chair, dtf_desk, dtf_dresser, dtf_monitor, dtf_night, dtf_sofa, dtf_table, dtf_toilet]


number_of_test = [];
number_of_train = [];
for index in range(10):
  #re ordering section to be in numerical order
  dtf_train, dtf_test = frames[index][frames[index]['split'] == 'train'], frames[index][frames[index]['split'] == 'test']
  together = [dtf_train, dtf_test]
  frames[index] = pd.concat(together, ignore_index=True)

  #keep in numerical order when splitting to avoid problem with runtime quitting
  train, test = train_test_split(frames[index], test_size=0.2, shuffle=False)
  number_of_test.append(test.shape[0])
  number_of_train.append(train.shape[0])
  test['split'] = 'test'
  train['split'] = 'train'
  tt = [train, test]
  frames[index] = pd.concat(tt, ignore_index=True)
  display(frames[index])

# Export dataset as 'modelnet10_split.csv'
modelnet10_split = pd.concat(frames, ignore_index=True)

modelnet10_split.to_csv('ModelNet10_Split.csv')
files.download('ModelNet10_Split.csv')
# Split into train set and test set
dtf_train, dtf_test = modelnet10_split[modelnet10_split['split'] == 'train'], modelnet10_split[modelnet10_split['split'] == 'test']
```

```python
import matplotlib.pyplot as plt

# Look at index 4 and 6, which demonstrate overlapping cases.
y1 = []
y2 = []

for index in range(10):
  y2.append(number_of_train[index])
  y1.append(number_of_train[index] + number_of_test[index])

# Colors: https://matplotlib.org/api/colors_api.html

colors = sns.color_palette("icefire", 1)
colors2 = sns.color_palette("crest", 1)

plt.figure(figsize=(10, 7))

print(classes)
print(y1)

plt.bar(classes, y1, label="Test", color=colors)
plt.bar(classes, y2, label="Train", color=colors2)

for index in range(10):
  print(classes[index] + ": " + str(y2[index] / y1[index]))

plt.plot()

plt.xlabel("Object")
plt.ylabel("Number of Meshes")
plt.title("Test / Train split of Meshes")
plt.legend()
plt.show()
```
